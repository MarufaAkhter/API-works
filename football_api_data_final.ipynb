## Football Data Analysis by rapid API

## Introduction: 

The data has been taken from rapid API. The link is given below:
https://rapidapi.com/api-sports/api/api-football-beta/

From there we took Top scorers within the year 2010 to 2020.


## Import libraries

```python
import json
import pandas as pd
from IPython.display import HTML
import os
import requests
from credentials import *
```

## Calling from API 

```python
url = "https://api-football-beta.p.rapidapi.com/players/topscorers"

result = []

for year in range(2010, 2020):
    querystring = {"season":str(year),"league":"39"}

    with open('football_api.json', 'r') as fi:
        headers = json.load(fi)
        
    response = requests.request("GET", url, headers=headers, params=querystring)
    result.append(response.json())

with open('file1.json', 'w') as file:
    json.dump(result, file)
```

## Importing the datasets

```python
with open('file1.json', 'rb') as f:
    data = json.load(f)
pd.json_normalize(data[10:], record_path =["response"], max_level=0)
```

```python
with open('file1.json', 'rb') as f:
    data = json.load(f)
data = pd.json_normalize(data[10:], record_path =["response"])

data["statistics"] = data.statistics.map(lambda x: x[0])
data2 = pd.json_normalize(data['statistics'])
data1 = data.loc[:,'player.id':]
data3 = pd.concat([data2,data1],axis=1)
data3
```

## Select the necessary columns to make a clean dataFrame

```python
df = data3[['team.name', 'team.logo', 'league.name',
       'league.country', 'league.logo', 'league.flag', 'league.season',
       'games.appearences', 'games.lineups', 'games.minutes', 'games.number',
       'games.position', 'games.rating', 'games.captain', 'substitutes.in',
       'substitutes.out', 'substitutes.bench', 'shots.total', 'shots.on',
       'goals.total', 'goals.conceded', 'goals.assists', 'goals.saves',
       'passes.total', 'passes.key', 'passes.accuracy', 'tackles.total',
       'tackles.blocks', 'tackles.interceptions', 'duels.total', 'duels.won',
       'dribbles.attempts', 'dribbles.success', 'dribbles.past', 'fouls.drawn',
       'fouls.committed', 'cards.yellow', 'cards.yellowred', 'cards.red',
       'penalty.won', 'penalty.commited', 'penalty.scored',  
           'player.firstname','player.lastname', 'player.age', 
       'player.birth.place', 'player.birth.country', 'player.nationality',
       'player.height', 'player.weight', 'player.injured', 'player.photo']]
df
```

## Parsing image columns to dataframe

```python
def path_to_image_html(path):
    return '<img src="'+ path + '" width="30" >'
df.loc[:,['player.photo', 'league.flag', 'team.logo', 'league.logo']] = df[['player.photo', 'league.flag', 'team.logo', 'league.logo']].apply(lambda x: path_to_image_html(x))
```
